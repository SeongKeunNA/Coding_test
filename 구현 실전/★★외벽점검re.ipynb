```python
from itertools import combinations
from itertools import combinations_with_replacement
def solution(n, weak, dist):
    answer = 0
    weak_copy = weak[:]
    idx_list = list(range(len(dist)))
    for num in range(1, len(dist)+1):
        possible = list(combinations(idx_list, num))
        for p in possible:
            start = list(combinations_with_replacement(list(range(n)), num))
            rotate = list(combinations_with_replacement(list(range(2)), num))
            for s in start:
                for r in rotate:
                    weak = weak_copy[:]
                    print('weak:',weak)
                    print('친구의 수:', num, '친구의 조합:', p, 'c친구들의 시작위치:', s, '친구들의 회전위치:',r)
                    for f in range(len(p)):
                        print('weak:',weak)
                        if dist[p[f]] >= n:
                            return num
                        
                        if r[f] == 0:
                            end = s[f] + dist[p[f]]
                            print('%d 번째 친구는 %d부터 %d까지 이동:' %(f+1, s[f], end))
                            for i in range(s[f], end+1):
                                print('i:',i,'i%n:',i%n)
                                if i%n in weak:
                                    weak.remove(i%n)
                                    
                        elif r[f] == 1:
                            end = s[f] - dist[p[f]] #s[f] = 1 end = 0 
                            print('%d 번째 친구는 %d부터 %d까지 이동:' %(f+1, s[f], end%n ))
                            if end < 0:
                                for i in range(end % n, n):
                                    if i%n in weak:
                                        weak.remove(i%n)
                                for i in range(s[f]+1):
                                    print('i:',i)
                                    if i in weak:
                                        weak.remove(i)
                                        
                            else:
                                for i in range(end, s[f]+1):
                                    print('i:',i)
                                    if i in weak:
                                        weak.remove(i)
    
                        if len(weak) == 0:
                            return num
    return -1
solution(12, [1,5,6,10], [1,2,3,4])
```

# 2차 답안 작성 전 생각

```python
""""""""""
1. 친구의 숫자를 1명부터 len(dist)까지 늘려가며,

2. 가장 이동 거리가 긴 num 명의 친구를 뽑는다.

3. 만약 len(weak) < num이라면 return num한다.

3. num명(ex dis2, dis3, dis7)의 친구들의 시작 위치 리스트는 취약 지점 중 num개를 순열로 뽑는다.

4. num명의 친구들의 회전 방향 리스트를 combinations_with_replacemnet를 이용해 구한다.

5. 가능한 시작위치와 가능한 회전 방향마다 반복하여:
    각각의 친구들을 각자의 이동가능 거리만큼 이동시키고 그 사이에 있는 취약점들을 제거한다.
    만약 취약점이 모두 제거됐으면 return num
    
6. return -1
""""""""""
```

```python
from itertools import permutations
from itertools import product
def solution(n, weak, dist):
    weak_copy = weak[:]
    for num in range(1, len(dist)+1):
        if len(weak) < num:
            return num
        friends = dist[-num:]
        start = list(permutations(weak, num)) 
        rotate = list(product([0,1], repeat = num))
        for s in start: 
            for r in rotate:
                weak = weak_copy[:]
                for f in range(len(friends)):
                    weak_ = weak[:]
                    if friends[f] >= n:
                        return num
                    
                    if r[f] == 0:
                        end = s[f] + friends[f]
                        if end < n:    
                            for w in weak_:
                                if s[f] <= w <= end:
                                    weak.remove(w)
                        else:
                            for w in weak_:
                                if s[f] <= w < n or 0 <= w <= end % n:
                                    weak.remove(w)
                    
                    elif r[f] == 1:
                        end = s[f] - friends[f]
                        if end < 0:
                            for w in weak_:
                                if end % n <= w < n or 0 <= w <= s[f]:
                                    weak.remove(w)

                        else:
                            for w in weak_:
                                if end <= w <= s[f]:
                                    weak.remove(w)
    
                    if len(weak) == 0:
                        return num
    return -1
solution(12, [1,5,6,10], [1,2,3,4])
```

```python
from itertools import permutations
def solution(n, weak, dist):
    for num in range(1, len(dist)+1):
        if len(weak) < num:
            return num
        friends = dist[-num:]
        start = list(permutations(weak, num)) 
        for s in start:
            fix = set()
            for f in range(len(friends)):
                last = list(set(weak) - fix)
                if friends[f] >= n:
                    return num
                end = s[f] + friends[f]
                if end < n:  
                    for w in last:
                        if s[f] <= w <= end:
                            fix.add(w)
                else:
                    for w in last:
                        if s[f] <= w < n or 0 <= w <= end % n:
                            fix.add(w)


                if len(fix) == len(weak):
                    return num
    return -1
```

```python
from itertools import permutations
s = time.time()
def solution(n, weak, dist):
    length = len(weak)
    for i in range(length):
        weak.append(weak[i]+n) 
    for c in range(1, len(dist)+1):
        friends = dist[-c:] 
        for start in range(length): 
            for fp in list(permutations(friends, c)): 
                count = 1 
                position = weak[start] + fp[count-1] 
                bk = 0
                for index in range(start, start+length): 
                    if position < weak[index]: 
                        count += 1 
                        if count > len(fp):
                            bk= 1
                            break
                        position = weak[index] + fp[count-1]
                if bk == 0:
                    return c
    return -1
print(solution(200, weak, dist))
print(time.time()-s)
```
