```python
# 내 답안
def solution(n, build_frame):
    answer = []
    def remove_ok(x,y,a):
            if a == 0:
                if [x, y+1, 0] not in answer and [x, y+1, 1] not in answer and [x-1, y+1, 1] not in answer:
                    return True
                
                elif [x, y+1, 0] in answer and [x, y+1, 1] not in answer and [x-1, y+1, 1] not in answer:
                    return False
                    
                elif [x, y+1, 1] in answer and [x-1, y+1, 1] not in answer:
                    if [x+1, y, 0] in answer:
                        return True
                    else:
                        return False
                elif [x, y+1, 1] not in answer and [x-1, y+1, 1] in answer:
                    if [x-1, y, 0] in answer:
                        return True
                    else:
                        return False
                elif [x, y+1, 1] in answer and [x-1, y+1, 1] in answer:
                    if ([x-2, y+1, 1] in answer or [x-1, y, 0] in answer) and ([x+1, y+1, 1] in answer or [x+1, y, 0] in answer):
                        return True
                    else:
                        return False
                    
            else:
                if [x-2,y,1] in answer and [x-1,y,1] in answer and [x-1,y-1,0] not in answer and [x,y-1,0] not in answer:
                    return False
                if [x+1,y,1] in answer and [x+2,y,1] in answer and [x+1,y-1,0] not in answer and [x+2,y-1,0] not in answer:
                    return False
                if [x,y,0] in answer:
                    if [x-1,y,1] not in answer and [x, y-1, 0] not in answer:
                        return False
                if [x+1,y,0] in answer:
                    if [x+1,y,1] not in answer and [x+1, y-1, 0] not in answer:
                        return False
                return True
                        
    def build_ok(x,y,a):
        if a == 0:
            if y == 0 or [x-1,y,1] in answer or [x, y-1, 0] in answer or [x, y, 1] in answer:
                return True
            else:
                return False
        else:
            if ([x,y-1,0] in answer or [x+1,y-1,0] in answer) or (  [x-1,y,1] in answer and [x+1,y,1] in answer ):
                return True
            else: 
                return False
            
    for x,y,a,b in build_frame:
        if (x < 0 or x > n or y < 0 or y > n) or (a == 1 and (y == 0 or x == n)) or (a == 0 and  y == n):
            continue
        
        elif b == 0:
            if remove_ok(x,y,a):
                answer.remove([x,y,a])
            else:
                continue       
        else:
            if build_ok(x,y,a):
                answer.append([x,y,a])
            else:
                continue
            
    answer.sort(key = lambda x: (x[0],x[1],x[2]))
    return answer
```
