```python
def solution(key, lock):
    M=len(key)
    N=len(lock)
    def r(key):
        rkey=[]
        subkey = []
        for i in range(0, M):
            for j in range(M-1, -1, -1):
                subkey.append(key[j][i])
            rkey.append(subkey)
            subkey = []
        return rkey
            
    def mr(key):
        rkey= [[0]* 1 for _ in range(M)]
        for i in range(M):
            for j in range(M-1):
                rkey[i].append(key[i][j])
        return rkey
    
    def ml(key):
        rkey = key
        for i in range(M):
            del(rkey[i][0])
            rkey.append(0)
        return rkey
    

    def mu(key):
        rkey = key[1:]
        rkey.append([0] * 3)
        return rkey
    
    def md(key):
        rkey =[[0]*3] 
        rkey.append(key[:M-1])
        return rkey
    
    def search(key):
        for i in range(M-N, N):
            for j in range(M-N, N):
                for k in range(i, i+M):
                    for l in range(j, j+M):
                        print(i, j, k, l)
                        if 0 <= k <= N-1 and 0 <= l <= N-1:
                            if lock[k][l] + key[k - i][l - j] == 1:
                                pass
                        else:
                            break
            return True
        return False
    mnum = 0
    nnum = 0
    for i in range(M):
        for j in range(M):
            mnum += key[i][j]
    for i in range(N):
        for j in range(N):
            nnum += lock[i][j]
    if mnum < (N**2-nnum):
        return False
        
    r_set=['k','1','2','3']
    move_set=['k','r','l','u','d']
    
    for r in r_set:
        if r == 'k':
            pass
        elif r == '1':
            key = r(key)

            
        elif r == '2':
            key = r(r(key))

        else:
            key = r(r(r(key)))        
        
        for m in move_set:
            if m == 'k':
                if search(key) == True:
                    return True
            elif m == 'r':
                key = mr(key)
                if search(key) ==True:
                    return True
            elif m == 'l':
                key = ml(key)
                if search(key) ==True:
                    return True
            elif m == 'u':
                key = mu(key)
                if search(key) ==True:
                    return True
            else:
                key = md(key)
                if search(key) ==True:
                    return True
    return False
```
