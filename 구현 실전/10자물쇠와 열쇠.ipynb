```python
def solution(key, lock):
    M=len(key)
    N=len(lock)
    def r1(key):
        rkey = [[]]
        subkey = []
        for i in range(0, M):
            for j in range(M-1, -1, -1):
                subkey.append(key[j][i])
            rkey.append(subkey)
            subkey = []
        return rkey
            
    def r2(key):
        rkey = [[]]
        subkey = []
        for i in range(M-1, -1, -1):
            for j in range(M-1, -1, -1):
                subkey.append(key[j][j])
            rkey.append(subkey)
            subkey = []
        return rkey
            
    def r3(key):
        rkey = [[]]
        subkey = []
        for i in range(M-1, -1, -1):
            for j in range(0, M):
                subkey.append(key[j][i])
            rkey.append(subkey)
            subkey = []
        return rkey
    
    def mr(key):
        rkey= [[0]* 1 for _ in range(M)]
        for i in range(M):
            for j in range(M-1):
                rkey[i].append(key[i][j])
        return rkey
    
    def ml(key):
        rkey=[]
        for i in range(M):
            for j in range(1,M):
                rkey.append(key[i][j])
            rkey.append(0)
        return rkey
    

    def mu(key):
        rkey = key[1:]
        rkey.append([0] * 3)
        return rkey
    
    def md(key):
        rkey =[[0]*3] 
        rkey.append(key[:M-1])
        return rkey
    
    def search(key):
        for i in range(0,N-M):
            for j in range(M,N):
                for q in range(M):
                    for w in range(M):
                        if key[q][w] + lock[i][j] == 1:
                            pass
                        else:
                            return False
        return True    

    mnum = 0
    nnum = 0
    for i in range(M):
        for j in range(M):
            mnum += key[i][j]
    for i in range(N):
        for j in range(N):
            nnum += lock[i][j]
    if mnum < (9-nnum):
        answer = False
    else:
        r_set=['k','1','2','3']
        move_set=['k','r','l','u','d']
    
    for r in r_set:
        if r == 'k':
            pass
        elif r == '1':
            key = r1(key)
            
        elif r == '2':
            key = r2(key)

        else:
            key = r3(key)
        
        for m in move_set:
            if m == 'k':
                if search(key) == True:
                    return True
            elif m == 'r':
                key = mr(key)
                if search(key) ==True:
                    return True
            elif m == 'l':
                key = ml(key)
                if search(key) ==True:
                    return True
            elif m == 'u':
                key = mu(key)
                if search(key) ==True:
                    return True
            elif m == 'd':
                key = md(key)
                if search(key) ==True:
                    return True
    return False
```
