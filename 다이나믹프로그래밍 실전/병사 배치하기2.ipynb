# 답지 초반 부분 참고 후 답안

```python
n = int(input())
data = list(map(int, input().split()))
a = [1] * n

for i in range(1, n):
    a[i] = a[i-1]
    if data[i] <= data[i-1]:
        a[i] += 1
print(n - a[n-1])
```

# 정답

```python
n = int(input())
array = list(map(int, input().split()))
# 순서를 뒤집어 '가장 긴 증가하는 부분 수열' 문제로 변환
array.reverse()

# 다이나믹 프로그래밍을 위한 1차원 DP 테이블 초기화
dp = [1] * n

# 가장 긴 증가하는 부분 수녕 (LIS) 알고리즘 수행
for i in range(1, n):
    for j in range(0, i):
        if array[j] < array[i]:
            dp[i] = max(dp[i], dp[j] + 1)
# 열외시켜야 하는 병사의 최소 수를 출력
print(n- max(dp))
```
