```python
n, l, r = map(int ,input().split())
land = []
for _ in range(n):
    land.append(list(map(int, input().split())))
    
visited = [[0 for _ in range(n)]for _ in range(n)]
union_sum = [[0 for _ in range(n)]for _ in range(n)]
union_num = [[0 for _ in range(n)]for _ in range(n)]
union_mean_list = land
dx = [0, 0, 1, -1]
dy = [1, -1, 0 ,0]
cnt = 0

def union(ux, uy, x, y, pre_num, visited, union_sum, union_num, ck=0):
    print('x:', x, 'y:', y, 'pre_num:', pre_num, 'land:', land, 'visited:', visited)
    print('union_mean_list:',union_mean_list, 'union_sum: ', union_sum, 'union_num:', union_num)
    if x < 0 or x >= n or y < 0 or y >= n or visited[x][y] != 0:
        return False
    
    if x == ux and y == uy and visited[x][y] == 0:
        visited[x][y] = (ux, uy)
        union_num += 1
        union_sum += land[x][y]
    elif l <= abs(land[x][y] - pre_num) <= r:
        ck = 1
        union_sum[ux][uy] += land[x][y]
        union_num[ux][uy] += 1
        visited[x][y] = (ux, uy)
        
    else:
        return False
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        union(ux, uy, nx, ny, land[x][y], visited, union_sum, union_num, 0)  
    return True
        
    if union_num[ux][uy] >=1:
        union_mean_list[ux][uy] = union_sum[ux][uy]  // union_num[ux][uy]
    
    return ck
 
def move_and_ck(land, move_ck):
    for ux in range(n):
        for uy in range(n):
            if union(ux, uy, ux, uy, land[ux][uy], visited, union_sum, union_num, 0):
                move_ck = 1
    print('land: ', land, 'union_mean_list:', union_mean_list, 'visited:', visited)
    for ux in range(n):
        for uy in range(n):
            land[ux][uy] = union_mean_list[visited[ux][uy][0]][visited[ux][uy][1]]
         
            
while True:
    if move_and_ck(land, 0) == False:
        print(cnt)
        break
        
    else:
        cnt += 1
        for i in range(n):
            for j in range(n):
                land[i][j] = union_mean_list[visited[i][j][0]][visited[i][j][0]]
```

```python
n, l, r = map(int ,input().split())
land = []
for _ in range(n):
    land.append(list(map(int, input().split())))
visited = [[0 for _ in range(n)]for _ in range(n)]
union_mean_list = [[0 for _ in range(n)]for _ in range(n)]
dx = [0, 0, 1, -1]
dy = [1, -1, 0 ,0]
cnt = 0

def union(ux, uy, x, y, pre_num, visited, union_mean):
    print('x:', x, 'y:', y, 'pre_num:', pre_num, 'land:', land, 'visited:', visited)
    print('union_mean_list:',union_mean_list, 'union_mean:', union_mean)
    if x < 0 or x >= n or y < 0 or y >= n or visited[x][y] != 0:
        return False
    
    if ux == x and uy == y:
        visited[x][y] = (ux, uy)
            
    elif l <= abs(land[x][y] - pre_num) <= r:
        visited[x][y] = (ux, uy)
        union_mean = (pre_num + land[x][y]) // 2
    
    else:
        return False
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        union(ux, uy, nx, ny, land[x][y], visited, union_mean)   
    union_mean_list[ux][uy] = union_mean
    return True
        
for ux in range(n):
    for uy in range(n):
        union(ux, uy, ux, uy, land[ux][uy], visited, union_mean) # union(0, 0, 0, 0, 50, visited, union_mean)    
for i in  range(n):
        for j in range(n):
            land[i][j] = union_mean_list[visited[i][j][0]][visited[i][j][1]] 
            
for x in range(n):
    for y in range(n):
        ck = []      
        for k in range(4):
            nx = x + dx[k]
            ny = y + dy[k]   
            if nx >= 0 and nx < n and ny >= 0 and ny < n:
                vn = visited[nx][ny]
                if  vn not in ck:    
                    ck.append(vn)
                    if l <= abs(land[x][y] - land[nx][ny]) <= r:
                        cnt += 1
print(cnt)
```
