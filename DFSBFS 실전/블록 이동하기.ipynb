# 내 답안

```python
robot = [[0,0],[0,1]]
dx = [0, 0, 1 ,-1]
dy = [1, -1, 0, 0]
answer = 1e9
board = [[0, 0, 0, 1, 1],[0, 0, 0, 1, 0],[0, 1, 0, 1, 1],[1, 1, 0, 0, 1],[0, 0, 0, 0, 0]]
n = 5
def move_check(i):
    global robot
    global board
    print('robot:', robot, 'board:',board)
    for r in robot:
        if r[0] + dx[i] < 0 or r[0] + dx[i] > n or r[1] + dy[i] < 0  or r[1] + dy[i] > n:
            return False
        
        if board[(r[0] + dx[i])][r[1] + dy[i]] == 1:
            return False
            
    for r in robot:
        if board[r[0]+dx[i]][r[1]+dy[i]] != 2:
            return True
    return False

def move(i):
    global robot
    global board
    for r in robot:
        r[0] += dx[i]
        r[1] += dy[i]
        board[r[0]][r[1]] = 2
    return True

def trans(i):
    if i == 0:
        j = 1
    elif i == 1:
        j = 0
    elif i == 2:
        j = 3
    else:
        j = 2
    return j
    
def rotate_check(clockwise, axis, state):
    global robot
    global board
    if robot[0][0] == robot[1][0]:
        state = 0   
    else:
        state = 1
        
    if clockwise == 0 and state == 0 and axis == 0:
        if robot[0][1] <= n-1:
            if board[robot[0][0]][(robot[0][1])+1] == 0 and board[robot[0][0]-1][(robot[0][1])+1] == 0:
                if board[robot[1][0]-1][robot[1][1]+1] != 2:
                    return True
        return False
    
    elif clockwise == 0 and state == 0 and axis == 1:
        if robot[0][1] >= 1:
             if board[robot[0][0]][(robot[0][1])-1] == 0 and board[robot[0][0]+1][(robot[0][1])-1] == 0:
                    if board[robot[0][0]+1][robot[0][1]-1] != 2:
                        return True
        return False
    
    
    elif clockwise == 0 and state == 1 and axis == 0:
        if robot[0][0] >= 1:
             if board[robot[0][0]-1][(robot[0][1])] == 0 and board[robot[0][0]-1][(robot[0][1])+1] == 0:
                if board[robot[1][0]-1][robot[1][1]-1] != 2:
                    return True
        return False
    
    elif clockwise == 0 and state == 1 and axis == 1:
        if robot[0][1] <= n-1:
             if board[(robot[axis][0])+1][(robot[axis][1])] == 0 and board[robot[axis][0]+1][(robot[axis][1])-1] == 0:
                    if board[robot[abs(axis-1)][0] + 1][robot[abs(axis-1)][1] + 1] != 2:
                        return True
        return False
    
    elif clockwise == 1 and state == 0 and axis == 0:
        if robot[0][1] >= 1:
            if board[(robot[axis][0])-1][(robot[axis][1])-1] == 0 and board[robot[axis][0]][(robot[axis][1])-1] == 0:
                if board[abs(axis-1)[0] - 1][abs(axis-1)[1] - 1] != 2:
                    return True
        return False
    
    elif clockwise == 1 and state == 0 and axis == 1:
        if robot[1][1] <= n-1:
             if board[(robot[axis][0])-1][(robot[axis][1]+1)] == 0 and board[robot[axis][0]][(robot[axis][1])+1] == 0:
                if board[robot[abs(axis-1)[0] + 1]][robot[abs(axis-1)[1] + 1]] != 2:
                    return True
        return False
    
    
    elif clockwise == 1 and state == 1 and axis == 0:
        if robot[0][0] <= n-1:
             if board[(robot[axis][0])+1][(robot[axis][1])] == 0 and board[robot[axis][0]+1][(robot[axis][1])+1] == 0:
                if board[robot[abs(axis-1)][0] + 1][robot[abs(axis-1)][1] - 1] != 2:
                    return True
                                
        return False
                                                                                   
    else:
        if robot[1][0] >= 1:
             if board[(robot[axis][0])-1][(robot[axis][1])-1] == 0 and board[robot[axis][0]-1][(robot[axis][1])] == 0:                                                   
                if board[robot[abs(axis-1)][0] - 1][robot[abs(axis-1)][1] + 1] != 2:
                    return True
        return False
                                                   
                                                   
                                                
def rotate(clockwise=0, axis=0):
    global robot
    global board
    if robot[0][0] == robot[1][0]:
        state = 0   
    else:
        state = 1
                                            
    if clockwise == 0 and state == 0 and axis == 0:
        robot[1][0] -= 1
        robot[1][1] += 1
        board[robot[1][0]][robot[1][1]] = 2
        return True
            
    elif clockwise == 0 and state == 0 and axis == 1:
        robot[0][0] += 1
        robot[0][1] -= 1
        board[robot[0][0]][robot[0][1]] = 2
        return True
    
    elif clockwise == 0 and state == 1 and axis == 0:
        robot[1][0] -= 1
        robot[1][1] -= 1
        board[robot[1][0]][robot[1][1]] = 2
        return True
    
    
    elif clockwise == 0 and state == 1 and axis == 1:
        robot[abs(axis-1)][0] += 1
        robot[abs(axis-1)][1] += 1
        board[robot[0][0]][robot[0][1]] = 2
        return True
    
    elif clockwise == 1 and state == 0 and axis == 0:
        robot[abs(axis-1)][0] -= 1
        robot[abs(axis-1)][1] -= 1
        board[robot[1][0]][robot[1][1]] = 2
        return True
    
    elif clockwise == 1 and state == 0 and axis == 1:
        robot[abs(axis-1)][0] += 1
        robot[abs(axis-1)][1] += 1
        board[robot[0][0]][robot[0][1]] = 2
        return True
    
    elif clockwise == 1 and state == 1 and axis == 0:
        robot[abs(axis-1)][0] += 1
        robot[abs(axis-1)][1] -= 1
        board[robot[1][0]][robot[1][1]] = 2
        return True
    
    else:
        robot[abs(axis-1)][0] -= 1
        robot[abs(axis-1)][1] += 1
        board[robot[0][0]][robot[0][1]] = 2
        return True
    return False


                    
def process(cnt):
    global robot
    global board
    for r in robot:
        if r[0] == n-1 and r[1] == n-1:
            answer = min(answer, cnt)
            return False
        
    for i in range(4):
        if move_check(i):
            move(i)
            cnt += 1
            process(cnt)
            j = trans(i)
            move(j)
            cnt -= 1

            
    for i in range(2):
        for j in range(2):
            for k in range(3):
                if rotate_check(i,j,_):
                    rotate(i,j)
                    cnt += 1
                    process(cnt)
                    rotate(abs(i-1),j)
                    cnt -= 1
                else:
                    break
process(0)
print(answer)
```
