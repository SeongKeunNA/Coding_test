# 내 답안

```python
from itertools import permutations
n = int(input())
A = list(map(int,input().split()))
oper_num = list(map(int,input().split()))
oper_list = []
oper = ['+','-','x','//']
for i in range(4):
    if oper_num[i] == 0:
        continue
    for _ in range(oper_num[i]):
        oper_list.append(oper[i])
oper_deque_list = list((permutations(oper_list,n-1)))
answer_list = [A[0]] * len(oper_deque_list)

for i in range(len(oper_deque_list)): # i = 0, 1
    for j in range(n-1): # 0, 1
        if oper_deque_list[i][j] == '+':
            answer_list[i] += A[j+1]
        elif oper_deque_list[i][j] == '-':
            answer_list[i] -= A[j+1]
        elif oper_deque_list[i][j] == 'x':
            answer_list[i] *=  A[j+1]        
        else:
            if answer_list[i] < 0 and j > 0:
                answer_list[i] = -(answer_list[i]) //  A[j+1]
                answer_list[i] = -(answer_list[i])
            else:
                answer_list[i] = answer_list[i] // A[j+1]
print(max(answer_list))
print(min(answer_list))
```

# 정답

```python
n = int(input())
# 연산을 수행하고자 하는 수 리스트
data = list(map(int, input().split()))
# 더하기, 빼기, 곱하기, 나누기 연산자 개수
add, sub, mul, div = map(int,input().split())

min_value = 1e9
max_value = -1e9

# 깊이 우선 탐색 메서드
def dfs(i, now):
    global min_value, max_value, add, sub, mul, div
    # 모든 연산자를 다 사용한 경우, 최솟값과 최댓값 업데이트
    if i == n:
        min_value = min(min_value, now)
        max_value = max(max_valuem now)
    else:
        # 각 연산자에 대하여 재귀적으로 수행
        if add > 0:
            add -= 1
            dfs(i + 1, now + data[i])
            add += 1
        if sub > 0:
            sub -= 1
            dfs(i + 1, now - data[i])
            sub += 1
        if mul > 0:
            mul -= 1
            dfs(i +1, now * data[i])
            mul += 1
        if div > 0:
            div -= 1
            dfs(i + 1, int(now / data[i]))
            div += 1
dfs(1, data[0])

print(max_value)
print(min_value)
```
