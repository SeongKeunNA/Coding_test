# 내 답안

```python
from itertools import permutations
n = int(input())
sequence = list(map(int, input().split()))
data = list(map(int, input().split()))
oper_all = []
operaters = ['+', '-', '*', '//']
oper_check = [1] * len(oper_all)

for i in range(4):
    for _ in range(data[i]):
        oper_all.append(operaters[i])
possible = set(permutations(oper_all, n-1))
min_value = 1e9
max_value = -1e9
for p in possible:
    val = sequence[0]
    for i in range(n-1):
        if p[i] == '+':
            val += sequence[i+1]
        elif p[i] == '-':
            val -= sequence[i+1]
        elif p[i] == '*':
            val *= sequence[i+1]
        elif val > 0:
                val //= sequence[i+1]
        else:
            val = -(-val // sequence[i+1])
    min_value = min(min_value, val)
    max_value = max(max_value, val)
print(max_value)
print(min_value)
```

# 답안 작성 후 생각

```python
""""""""""
dfs를 이용해 풀어보려고 했는데 생각보다 간단하지 않았다. 우선 연구소 문제와 달리 맵의 형식으로 되어 있지도 않고,

단순히 어느 위치에 벽을 세울지가 중요한 게 아니라 어떤 순서로 뽑아 낼지가 중요하기 때문이다.

게다가 같은 연사자도 있어서 조금 복잡해졌다.

우선 permutations으로 구현해 백준에서 정답을 기록했다.

내일 dfs로 풀어보자.
""""""""""
```

```python
from collections import deque
n = int(input())
sequence = list(map(int, input().split()))
oper = list(map(int, input().split())) # 1 0 1 0
q = deque([]) # (1000, 1, 12) 
cnt = 0
val = sequence[0] # 3
min_value = 1e9
max_value = -1e9
def q_add(oper, cnt, val, q):
    for i in range(4): 
        if oper[i] > 0:
            new_oper = oper[:]
            new_oper[i] -= 1 # 0010
            new_val = val
            if i == 0:
                new_val += sequence[cnt+1]
            elif i == 1:
                new_val -= sequence[cnt+1]
            elif i == 2:
                new_val *= sequence[cnt+1]
            elif val >= 0:
                new_val //= sequence[cnt+1]
            else:
                new_val =-(-val // sequence[cnt+1])
            q.append((new_oper, cnt+1, new_val))
            
q_add(oper,cnt, val, q)
while q:
    oper, cnt, val = q.popleft() #(1000,1,12) (0000, 2, 17)
    if cnt == n-1:
        min_value = min(min_value, val)
        max_value = max(max_value, val)
        
    else:
        q_add(oper,cnt, val, q)

print(max_value)
print(min_value)
```

# 정답

```python
n = int(input())
data = list(map(int, input().split()))
add, sub, mul, div = map(int, input().split())

min_value = 1e9
max_value = -1e9

def dfs(i, now):
    global min_value, max_value, add, sub, mul, div
    if i == n:
        min_value = min(min_value, now)
        max_value = max(max_value, now)

    else:
        if add > 0:
            add -= 1
            dfs(i + 1, now + data[i])
            add += 1
        if sub > 0:
            sub -= 1
            dfs(i + 1, now - data[i])
            sub += 1
        if mul > 0:
            mul -= 1
            dfs(i + 1, now * data[i])
            mul += 1
        if div > 0:
            div -= 1
            dfs(i + 1, int(now / data[i]))
            div += 1
dfs(1, data[0])

print(max_value, min_value, sep = '\n')
```
