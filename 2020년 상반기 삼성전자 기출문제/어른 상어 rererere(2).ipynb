```python
n, m, k = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]
smell = [[0] * n for _ in range(n)]
for i in range(n):
    for j in range(n):
        if graph[i][j] > 0:
            smell[i][j] = [graph[i][j], k]
        else:
            smell[i][j] = [0, 0]

directions = list(map(int, input().split()))
priorities = [[list(map(int, input().split())) for _ in range(4)] for _ in range(m)]
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def move_sharks():
    temp = [[0] * n for _ in range(n)]
    for x in range(n):
        for y in range(n):
            if graph[x][y] > 0:
                for i in priorities[graph[x][y] - 1][directions[graph[x][y] - 1] - 1]:
                    nx = x + dx[i - 1]
                    ny = y + dy[i - 1]       
                    if 0 <= nx < n and 0 <= ny < n:                    
                        if smell[nx][ny][1] == 0:
                            directions[graph[x][y] - 1] = i
                            if temp[nx][ny] == 0:
                                temp[nx][ny] = graph[x][y]
                            else:
                                temp[nx][ny] = min(temp[nx][ny], graph[x][y])
                            break
                else:
                    for j in priorities[graph[x][y] - 1][directions[graph[x][y] - 1] - 1]:
                        nx = x + dx[j - 1]
                        ny = y + dy[j - 1]       
                        if 0 <= nx < n and 0 <= ny < n:     
                            if smell[nx][ny][0] == graph[x][y]:
                                directions[graph[x][y] - 1] = j
                                temp[nx][ny] = graph[x][y]
                                break
    return temp

def process_smell():
    for i in range(n):
        for j in range(n):
            if graph[i][j] > 0:
                smell[i][j] = [graph[i][j], k]
            elif smell[i][j][1] > 0:
                smell[i][j][1] -= 1
                
def count_sharks():
    count = 0
    for i in range(n):
        for j in range(n):
            if graph[i][j] > 0:
                count += 1
                if count > 1:
                    return False
    return True
                
total_count = 0
while True:
    if total_count <= 27:
        for g in smell: print(g)
        print('-----')
    if count_sharks():
        print(total_count)
        break
    elif total_count >= 1000:
        print(-1)
        break
    graph = move_sharks()
    process_smell()
    total_count += 1
```
