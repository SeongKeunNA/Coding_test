```python
n, m, k = map(int, input().split())
array = [[None] * n for _ in range(n)]
for i in range(n):
    data = list(map(int, input().split()))
    for j in range(n):
        if data[j] > 0:
            array[i][j] = [1, data[j] - 1, k]
        else:
            array[i][j] = [0, -1, 0]
now_d = list(map(int, input().split()))
priorities = [[] for _ in range(m)]
for i in range(m):
    for j in range(4):
        priorities[i].append(list(map(int, input().split())))        
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def move(x, y):
    num = array[x][y][1]
    for i in priorities[num][now_d[num] - 1]:
        nx = x + dx[i - 1]
        ny = y + dy[i - 1]
        if 0 <= nx < n and 0 <= ny < n:
            if array[nx][ny][2] == 0:
                return (nx, ny, i)
    else:
        for i in priorities[num][now_d[num] - 1]:
            nx = x + dx[i - 1]
            ny = y + dy[i - 1]
            if 0 <= nx < n and 0 <= ny < n:
                if array[nx][ny][1] == num:
                    return (nx, ny, i)
                
def find(num):
    for x in range(n):
        for y in range(n):
            if array[x][y][0] == 1 and array[x][y][1] == num:
                return x, y

            
def process():
    total_count = 0
    while True:   
        if total_count <= 5:
            for a in array:
                print(a)
        total_count += 1
        temp = [[[] for _ in range(n)] for _ in range(n)]
        for i in range(m):
            at = find(i)
            if at == None:
                continue
            x, y = at
            nx, ny, i = move(x, y)
            array[x][y][0] = 0              
            now_d[array[x][y][1]] = i
            temp[nx][ny].append((array[x][y][1], x, y))
        count = 0
        for i in range(n):
            for j in range(n):
                if len(temp[i][j]) >= 1:
                    temp[i][j].sort()
                    best = temp[i][j][0]
                    array[i][j] = [1, best[0], k]
                    count += 1            
        if count == 1:
            return total_count
        elif total_count > 1000:
            return -1
        for x in range(n):
            for y in range(n):
                if array[x][y][0] == 0 and array[x][y][2] > 0:
                    array[x][y][2] -= 1
process()       
```

# 정답

```python
n, m, k = map(int, input().split())

array = []
for i in range(n):
    array.append(list(map(int, input().split())))
    
directions = list(map(int, input().split()))

smell =  [[[0, 0]] * n for _ in range(n)] ###

priorities = [[] for _ in range(m)]
for i in range(m):
    for j in range(4):
        priorities[i].append(list(map(int, input().split())))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1 ,1]

def update_smell():###
    for i in range(n):
        for j in range(n):
            if smell[i][j][1] > 0:
                smell[i][j][1] -= 1
            if array[i][j] != 0:
                smell[i][j] = [array[i][j], k]
def move(): ###
    new_array = [[0] * n for _ in range(n)]
    for x in range(n):
        for y in range(n):
            if array[x][y] != 0:
                direction = directions[array[x][y] - 1]
                found = False
                for index in range(4):
                    nx = x + dx[priorities[array[x][y] - 1][direction - 1][index] - 1]
                    ny = y + dy[priorities[array[x][y] - 1][direction - 1][index] - 1]
                    if 0 <= nx < n and 0 <= ny < n:
                        if smell[nx][ny][1] == 0:
                            directions[array[x][y] - 1] = priorities[array[x][y] - 1][direction - 1][index]
                            if new_array[nx][ny] == 0:
                                new_array[nx][ny] = array[x][y]
                            else:
                                new_array[nx][ny] = min(new_array[nx][ny], array[x][y])
                            found = True
                            break
                if found:
                    continue
                for index in range(4):
                    nx = x + dx[priorities[array[x][y] - 1][direction - 1][index] - 1]
                    ny = y + dy[priorities[array[x][y] - 1][direction - 1][index] - 1]
                    if 0 <= nx < n and 0 <= ny < n:
                        if smell[nx][ny][0] == array[x][y]:
                            directions[array[x][y] - 1] = priorities[array[x][y] - 1][direction - 1][index]
                            new_array[nx][ny] = array[x][y]
                            break
    return new_array

time = 0
while True:
    update_smell()
    new_array = move()
    array = new_array
    time += 1
    
    check = True
    for i in range(n):
        for j in range(n):
            if array[i][j] > 1:
                check = False
    if check:
        print(time)
        break
    if time >= 1000:
        print(-1)
        break
```
