# 내 답안(이진 탐색 구현)

```python
n, x = map(int, input().split())
array = list(map(int, input().split()))
def binary_search_left(array, x):
    min_idx = -1
    start = 0
    end = len(array) - 1
    while start <= end:
        mid = (start + end) // 2
        if array[mid] >= x:
            min_idx = mid
            end = mid - 1
        elif array[mid] < x:
            start = mid + 1
    return min_idx

def binary_search_right(array, x):
    max_idx = -1
    start = 0
    end = len(array) - 1
    while start <= end:
        mid = (start + end) // 2
        if array[mid] > x:
            end = mid - 1
        elif array[mid] <= x:
            max_idx = mid
            start = mid + 1
    return max_idx

left_idx = binary_search_left(array, x)
right_idx = binary_search_right(array, x)
if left_idx == -1 or right_idx == -1:
    print(-1)
else:
    print(right_idx - left_idx + 1)
```
