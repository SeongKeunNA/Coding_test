```python
def bsearch(query):
    val = 0
    p = 0
    if query[0] == '?' and query[-1] != '?':
        val = 0
        start, end = 1, len(query) - 1        
    elif query[0] != '?' and query[-1] == '?':
        val = 1
    
    while start <= end:
        mid = (start + end) // 2
        if query[mid] == '?':
            if val == 0:
                start = mid + 1
            else:
                end = mid - 1
        else:
            p = mid
            if val == 0:
                end = mid - 1
            else:
                start = mid + 1
    if val ==0:
        s, e = p,len(query)
    else:
        s, e = 0,p+1
    return [s,e]

def num_check(q, words):
    start, end = 0, len(words)-1
    while start <= end:
        mid = (start + end) // 2
        if len(words[mid]) == len(q):
            start, end = 0, mid
            min_index = mid
            max_index = mid
            while start <= end:
                mid = (start + end) // 2
                if len(words[mid]) < len(q):
                    start = mid + 1
                else:
                    end = mid - 1
                    min_index = min(mid, min_index)
            start, end = min_index, len(words)-1

            while start <= end:  
                mid = (start + end) // 2
                if len(words[mid]) > len(q):
                    end = mid - 1
                else:
                    start = mid + 1
                    max_index = max(mid, max_index)
            return words[min_index:max_index+1]
        
        elif len(words[mid]) > len(q):
            end = mid -1
        else:
            start = mid + 1
    return []
    
def possible_words(words, queries):
    queries_ = sorted(queries, key = lambda x:len(x))
    start, end = 0, len(words)-1
    min_index = 1e9
    max_index = -1e9
    while start <= end:
        mid = (start + end) // 2
        if len(words[mid]) < len(queries_[0]):
            start = mid + 1
        else:
            end = mid - 1
            min_index = min(mid, min_index)
    start, end = min_index, len(words)-1

    while start <= end:  
        mid = (start + end) // 2
        if len(words[mid]) > len(queries_[-1]):
            end = mid - 1
        else:
            start = mid + 1
            max_index = max(mid, max_index)
    return words[min_index:max_index+1]

def solution(words, queries):
    answer = []

    words.sort(key = lambda x:len(x))
    words = possible_words(words, queries)
    for q, i in enumerate (queries):
        print((q, i))
        
        
    queries.sort(key = lambda x:len(x))
    
    for query in queries:
        cnt = 0     
        val = 0
        if bsearch(query) == False:
            cnt += len(num_check(query,words))
        else:
            s,e = bsearch(query)
            for word in num_check(query, words):
                break_ = 0
                for i in range(s, e):
                    if word[i] != query[i]:
                        break_ = 1
                        break
                if break_ == 0:
                    cnt += 1             
        answer.append(cnt)  
    return answer   
```

# 3차 답안

```python
def bsearch(query):
    val = 0
    p = 0
    start, end = 0, 0
    if query[0] == '?' and query[-1] != '?':
        val = 0
        start, end = 1, len(query) - 1        
    elif query[0] != '?' and query[-1] == '?':
        val = 1
        start, end = 0, len(query) - 2

    while start <= end:
        mid = (start + end) // 2
        if query[mid] == '?':
            if val == 0:
                start = mid + 1
            else:
                end = mid - 1
        else:
            p = mid
            if val == 0:
                end = mid - 1
            else:
                start = mid + 1
    if val ==0:
        s, e = p,len(query)
    else:
        s, e = 0,p+1
    return [s,e]

def solution(words, queries):
    answer = []
    answer_array = []
    data=[]
    for i in range(len(words)):
        data.append((words[i],len(words[i])))
    max_value = max(data, key = lambda x:x[1])
    w_array = [[] for i in range(max_value[1] +1)]
    for i in range(len(words)): # i = 0,1,2,3,4,5
        w_array[data[i][1]].append(data[i][0]) #data = [('frodo', 5), ('front', 5), ('frost', 5), ('frozen', 6), ('frame', 5), ('kakao', 5)]  
    data = []    
    for i,q in enumerate(queries):
        data.append((i, len(q), q))
    max_value = max(data, key = lambda x:x[1])
    q_array = [[] for i in range(max_value[1] + 1)]
    
    
    for i in range(len(queries)): # i = 0,1,2,3,4
        q_array[data[i][1]].append({data[i][0]:data[i][2]})
    print(q_array)
    for i in range(len(q_array)):
        if len(q_array[i]) == 0:
            continue
            
        else:
            for q in q_array[i]:
                cnt = 0
                string = list(q.values())[0]
                if len(w_array[i]) != 0:
                    if string[0] == '?' and string[-1] == '?':
                        cnt += len(w_array[i])
                    else:
                        s, e = bsearch(string)
                        for word in w_array[i]:
                            break_ = 0
                            for j in range(s, e):
                                if word[j] != string[j]:
                                    break_ = 1
                                    break
                            if break_ == 0:
                                cnt += 1
                answer_array.append((list(q.keys())[0] ,cnt))
                cnt = 0
    answer_array.sort()
    for i in answer_array:
        answer.append(i[1])
    return answer
```
